```python
p, v, i, t, C = symbols("p v i t C")
C = 4e-3
A, B = symbols("A B")
v = A*exp(-100*t) + B*exp(-600*t)

# 1)

#V(0+)时刻，电流和电压关系
eq1 = diff(C*v, t).subs(t, 0) - 2

#V(0-)时刻
eq2 = v.subs(t, 0)-50
solve([eq1, eq2], [A, B])
```

```python
# 2) (C*V^2)/2
(C*v**2/2).subs({A:61, B:-11, t:0})
```

```python
def parallel(*args):
    ret = 0
    for arg in args:
        ret = 1/arg + ret
    return simplify(1/ret)

parallel(20, 60) + parallel(30, 70)
```
