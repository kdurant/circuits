```python
s = symbols("s")
Vi = symbols("Vi")
ZR1 = 2
ZR2 = 5
ZC1 = 1/(0.1*s)
ZC2 = 1/(0.2*s)

Vo = parallel(ZC1, ZR2+ZC2)/(ZR1 + parallel(ZC1, ZR2+ZC2)) * (ZC2/(ZR2+ZC2)) * Vi
simplify(Vo/Vi)

numer(simplify(Vo/Vi))   # 分子
denom(simplify(Vo/Vi))   # 分母
```

![](./problems/14-04.png)

```python
R,L,C, Vi, Vo,s = symbols("R,L,C, Vi, Vo,s")

# (a)
eq1 = (Vi-Vo)/(s*L) - Vo/(1/(s*C)) - Vo/R
ans = solve(eq1, Vo)
ans = ans[0].subs(s, I*w)
pprint(ans)
```

```python
# (b)
Vo = Vi*(R+I*w*L)/((R+I*w*L) + 1/(I*w*C))
Vo = cancel(Vo)
pprint(Vo)
```

```python
# b
R,L,C, Vs, Vo,s = symbols("R,L,C, Vs, Vo,s")

Za = s*L + 1/(1/R + s*C)
Zo = 1/(1/R + s*C)
ans = Zo/Za
ans = simplify(ans)
# ans = expand(ans)
pprint(ans)
```

```python
from sympy.physics.control.lti import TransferFunction
from sympy.physics.control.control_plots import bode_plot
s = symbols("s")

tf1 = TransferFunction(10, (1+s)*(10+s), s)
# tf1 = TransferFunction(1*s**2 + 0.1*s + 7.5, 1*s**4 + 0.12*s**3 + 9*s**2, s)
bode_plot(tf1)   
```

![](./problems/14-11.png)

```python
H = 0.2*(10+s)/(s*(2+s))
simplify(H)

tf1 = TransferFunction(numer(H), denom(H) , s)
bode_plot(tf1)  
```
